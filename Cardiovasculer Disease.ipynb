```python
##Dens plot
def dens_plot(features, class_var):
    sns.set_style("whitegrid")  # Correct function name
    g = sns.FacetGrid(df, hue=class_var)  # Data passed directly to FacetGrid
    g.map(sns.kdeplot, features, fill=True)  # Changed to kdeplot (for density) with fill=True
    g.add_legend()

# Call the function
dens_plot("age", "cholesterol")

# Show the plot
plt.show()
```

```python
##Split the Dataset

from sklearn.model_selection import train_test_split
x = df.iloc[:,:-1]
x.head()
```

```python
y = df.iloc[:,11]
y.head()
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size = 2,random_state =42)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
##PCA(Principal component Analysis)

from sklearn.decomposition import PCA
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x)
x_pca[0]
```

```python
pca.explained_variance_ratio_
```

```python
##KNN

from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import accuracy_score


knn_classifier = KNeighborsClassifier(n_neighbors=3) 
knn_classifier.fit(x_train, y_train)  


y_pred = knn_classifier.predict(x_test)


accuracy = accuracy_score(y_test, y_pred)


print(f"Accuracy: {accuracy}")
```

```python
##Naive Bayes classifier

from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
# GaussianNB
G_classifier = GaussianNB()
G_classifier.fit(x_train, y_train)
GaussianNB()
predictions_G = G_classifier.predict(x_test)
accuracy = metrics.accuracy_score(y_test, predictions_G)
accuracy
```

```python
# BernoulliNB
B_classifier = BernoulliNB()
B_classifier.fit(x_train, y_train)
BernoulliNB()
predictions_B = B_classifier.predict(x_test)
accuracy_B = metrics.accuracy_score(y_test, predictions_B)
accuracy_B
```

```python
## Decision Tree
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
DecisionTreeClassifier()
y_pred = clf.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
## Random Forest

from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators=100,
random_state=42)
rf_classifier.fit(x_train, y_train)
RandomForestClassifier(random_state=42)
y_pred = rf_classifier.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
## Boosting Algorithm

from sklearn.ensemble import AdaBoostClassifier
base_classifier = DecisionTreeClassifier(max_depth=3)
adaboost_classifier = AdaBoostClassifier(base_classifier,
n_estimators=50, random_state=42)
adaboost_classifier.fit(x_train, y_train)
AdaBoostClassifier(estimator=DecisionTreeClassifier(max_depth=1),
random_state=42)
y_pred = adaboost_classifier.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
accuracy
```

```python
## SVM
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)

```

```python
svm_classifier = SVC(kernel='linear', C=1.0)
svm_classifier.fit(x_train, y_train)
SVC(kernel='linear')
y_pred = svm_classifier.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
## Logistic Regression
from sklearn import linear_model

# Create an instance of the LogisticRegression model
lrg = linear_model.LogisticRegression()

# Fit the model with the training data
lrg.fit(x_train, y_train)

# Evaluate the model using the test data
score = lrg.score(x_test, y_test)


```
